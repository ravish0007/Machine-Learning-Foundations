```python
X=np.array(x)
Y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
def hypothesis(a,x,b):
    return a*x+b
```

```python
def error(a,x,b,y):
    e=0
    m=len(y)
    
    for i in range(m):
        e+=np.power((hypothesis(a,x[i],b)-y[i]),2)
        
    return (1/(2*m))*e    
```

```python
def step_gradient(a,x,b,y,learning_rate):
    grad_a=0
    grad_b=0
    m=len(x)
    
    for i in range(m):
        grad_a+=(1/m)*(hypothesis(a,x[i],b)-y[i])*x[i]
        grad_b+=(1/m)*(hypothesis(a,x[i],b)-y[i])
        
    a=a-grad_a*learning_rate
    b=b-grad_b*learning_rate
    
    return a,b
```

```python
def descent(init_a,x,init_b,y,learning_rate,iterations):
    a=init_a
    b=init_b
    
    for i in range(iterations):
        e=error(a,x,b,y)
        
        if i%1000==0:
            print(f"error:{np.sqrt(e)},a:{a},b:{b}")
                  
        a,b=step_gradient(a,x,b,y,learning_rate)
                  
    return a,b             
```

```python
a=0.6424195947464778
b=0.058457660513689294
learning_rate=0.001
iterations=10000
```

```python
final_a,final_b=descent(a,X_train,b,Y_train,learning_rate,iterations)
```

```python
plt.plot(X_train,Y_train,'r.',X_train,hypothesis(a,X_train,b),'b.',
        X_train,hypothesis(final_a,X_train,final_b),'g')
```
