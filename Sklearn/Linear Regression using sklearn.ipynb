```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
%matplotlib inline
```

```python
boston=load_boston()
```

```python
boston.keys()
```

```python
print(boston.DESCR)
```

```python
features=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
target=pd.DataFrame(boston.target,columns=['Target'])
```

```python
data=pd.concat([features,target],axis=1)
```

```python
data
```

```python
correlation=data.corr('pearson')
correlation
```

```python
abs(correlation['Target']).sort_values()
```

```python
X=np.array(data['RM'])
Y=np.array(data['Target'])
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X=X.reshape(-1,1)
Y=Y.reshape(-1,1)
```

```python
x=scaler.fit_transform(X)
y=scaler.fit_transform(Y)
```

```python
plt.plot(x,y,'r.')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor =LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
for i in zip(y_pred,y_test):
    print(i)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,regressor.predict(x_train),'b')

plt.xlabel("RM")
plt.ylabel("target")
plt.title("Liner Regression (train)")
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test,regressor.predict(x_test),'b')

plt.xlabel("RM")
plt.ylabel("target")
plt.title("Liner Regression (test)")
```

```python
r2_score(y_test,y_pred)
```
